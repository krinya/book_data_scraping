```python
def get_one_book_link(ISBN, searchString =  'https://www.hebban.nl/Search?q=', sleepTime = 1):
    
    starting_link_to_scrape = str(str(searchString) + str(ISBN))
    
    isbns = []
    book_link = []
    
    driver.get(starting_link_to_scrape)
    time.sleep(sleepTime)
    base_source = driver.page_source
    base_soup_book = BeautifulSoup(base_source, 'html.parser')
    
    all_book_link_on_page = base_soup_book.find_all('a', href=True, itemprop = "name")
    
    if len(all_book_link_on_page) > 0:
        for one_book_link in all_book_link_on_page:
            isbns.append(ISBN)
            book_link.append(one_book_link['href'])
    else:
        isbns.append(ISBN)
        book_link.append('NoLinkFound')
    
    d = {'isbn': isbns, 'book_link': book_link}
    df = pd.DataFrame(data=d)
    df = df.drop_duplicates()
    df['row_number'] = range(1, len(df) + 1)
    
    return(df)
```

```python
def get_stars(link_to_book, get_html = True, html_input = ''):
    
    if get_html == True:
        driver.get(link_to_book)
        time.sleep(1)
        base_source = driver.page_source
        base_soup_book = BeautifulSoup(base_source, 'html.parser')
    else:
        base_soup_book = html_input

    five_star_number = 0
    four_star_number = 0
    three_star_number = 0
    two_star_number = 0
    one_star_number = 0

    book_stars = base_soup_book.find_all(class_='yf ratings')

    
    for bs in book_stars:

        rs_numbers = bs.find_all(class_='rating-stars')
        rs_amounts = bs.find_all(class_='bar-amount')       

        for rs in range(0, 100):
            
            try:
                if '5' in rs_numbers[rs].get_text().strip():
                    five_star_number = rs_amounts[rs].get_text().strip()
                    
            except:
                ("Also here")
                continue

            try:
                if '4' in rs_numbers[rs].get_text().strip():
                    four_star_number = rs_amounts[rs].get_text().strip()
            except:
                continue

            try:
                if '3' in rs_numbers[rs].get_text().strip():
                    three_star_number = rs_amounts[rs].get_text().strip()
            except:
                continue

            try:
                if '2' in rs_numbers[rs].get_text().strip():
                    two_star_number = rs_amounts[rs].get_text().strip()
            except:
                continue

            try:
                if '1' in rs_numbers[rs].get_text().strip():
                    one_star_number = rs_amounts[rs].get_text().strip()
            except:
                continue
    return(five_star_number, four_star_number, three_star_number, two_star_number, one_star_number)
```

```python
def get_stat_below(link_to_book, get_html = True, html_input = ''):
    
    if get_html == True:
        driver.get(link_to_book)
        time.sleep(1)
        base_source = driver.page_source
        base_soup_book = BeautifulSoup(base_source, 'html.parser')
    else:
        base_soup_book = html_input

    book_stats = base_soup_book.find_all(class_='details clearfix yf-work-article-details work-article__details')

    book_genre = []
    book_serie = []
    book_aantal_pagina = []
    book_uitgeverij = []
    book_eerste_editie = []
    book_laaste_editi = []
    book_isbn = []

    for base_stats in book_stats:

        book_stats = base_soup_book.find_all(class_='')

        for stats in book_stats:

            try:
                stats_s = stats.get_text()
                if ("Genre" or "Aantal pagina" or "Uitgeverij" or "Eerste editie" or "Laatste editie" or "ISBN")  in stats_s:

                    dt = stats.find_all("dt")
                    dd = stats.find_all('dd')

                    for i in range(0,100):

                        try:

                            if 'Genre' in dt[i].get_text():                       
                                for value in dd[i]: 
                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_genre.append(value)
                                        except:
                                            book_genre.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue
                            
                        try:

                            if 'Serie' in dt[i].get_text():                       

                                for value in dd[i]:
                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_serie.append(value)
                                        except:
                                            book_serie.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue

                        try:

                            if 'Aantal pagina' in dt[i].get_text():                       

                                for value in dd[i]: 

                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_aantal_pagina.append(value)
                                        except:
                                            book_aantal_pagina.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue

                        try:

                            if 'Uitgeverij' in dt[i].get_text():                       

                                for value in dd[i]:
                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_uitgeverij.append(value)
                                        except:
                                            book_uitgeverij.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue
                            
                        

                        try:

                            if 'Eerste editie' in dt[i].get_text():                       
                                for value in dd[i]: 

                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_eerste_editie.append(value)
                                        except:
                                            book_eerste_editie.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue

                        try:
                            if 'Laatste editie' in dt[i].get_text():                       

                                for value in dd[i]: 

                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_laaste_editi.append(value)
                                        except:
                                            book_laaste_editi.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue

                        try:

                            if 'ISBN' in dt[i].get_text():                       

                                for value in dd[i]: 

                                    try:
                                        try:
                                            value = value.get_text(strip=True)
                                            book_isbn.append(value)
                                        except:
                                            book_isbn.append(value.strip())

                                    except:
                                        continue

                        except:
                            continue


                    try:
                        book_genre = list(filter(len, book_genre))
                    except:
                        continue
                    try: 
                        book_aantal_pagina = list(filter(len, book_aantal_pagina))
                    except:
                        continue
                    try: 
                        book_uitgeverij = list(filter(len, book_uitgeverij))
                    except:
                        continue
                    try: 
                        book_eerste_editie = list(filter(len, book_eerste_editie))
                    except:
                        continue
                    try: 
                        book_laaste_editi = list(filter(len, book_laaste_editi))
                    except:
                        continue
                    try: 
                        book_isbn = list(filter(len, book_isbn))
                    except:
                        continue
                    try:
                        i = book_isbn.index(",")
                        del book_isbn[i]
                    except ValueError:
                        continue



            except:
                continue
                
        return(book_genre, book_serie, book_aantal_pagina, book_uitgeverij, book_eerste_editie, book_laaste_editi, book_isbn)
```

```python
def get_stat_above(link_to_book, get_html = True, html_input = ''):
    
    if get_html == True:
        driver.get(link_to_book)
        time.sleep(1)
        base_source = driver.page_source
        base_soup_book = BeautifulSoup(base_source, 'html.parser')
    else:
        base_soup_book = html_input

    book_ranks = []
    book_gelezens = []
    book_aanhetlezens = []
    book_wiliklezens = []
    book_favoriets = []
    book_recensiess = []

    book_stats = base_soup_book.find_all(class_='work-stats')

    book_rank = []
    book_gelezen = []
    book_aanhetlezen = []
    book_wiliklezen = []
    book_favoriet = []
    book_recensies = []

    for stats in book_stats:

        stat_label = stats.find(class_='work-stats-label')
        stat_value = stats.find(class_='work-stats-value')

        try:
            stat_label_s = stat_label.get_text()
            stat_value_s = stat_value.get_text()
        except:
            continue


        if stat_label_s == 'Rank':
            book_rank.append(stat_value_s)

        if stat_label_s == 'Gelezen':
            book_gelezen.append(stat_value_s)

        if stat_label_s == 'Aan het lezen':
            book_aanhetlezen.append(stat_value_s)

        if stat_label_s == 'Wil ik lezen':
            book_wiliklezen.append(stat_value_s)

        if stat_label_s == 'Favoriet':
            book_favoriet.append(stat_value_s)

        if stat_label_s == 'Recensies':
            book_recensies.append(stat_value_s)

    if not book_rank:
        book_rank.append('-1')

    if not book_gelezen:
        book_gelezen.append('-1')

    if not book_aanhetlezen:
        book_aanhetlezen.append('-1')

    if not book_wiliklezen:
        book_wiliklezen.append('-1')

    if not book_favoriet:
        book_favoriet.append('-1')

    if not book_rank:
        book_recensies.append('-1')

    book_ranks.append(book_rank[0])
    book_gelezens.append(book_gelezen[0])
    book_aanhetlezens.append(book_aanhetlezen[0])
    book_wiliklezens.append(book_wiliklezen[0])
    book_favoriets.append(book_favoriet[0])
    book_recensiess.append(book_recensies[0])
    
    return(book_ranks, book_gelezens, book_aanhetlezens, book_wiliklezens, book_favoriets, book_recensiess)
```

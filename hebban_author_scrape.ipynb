```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from multiprocessing import Pool
import os
import seaborn as sns
import pandas as pd
import re as re
```

```python
# Import webdriver for automation
from selenium import webdriver
```

```python
# Importing the ChromeDriver class for handling webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
starting_link_to_scrape = "https://www.hebban.nl/auteurs#total=4000"
driver.get(starting_link_to_scrape)
time.sleep(60)
base_source = driver.page_source
time.sleep(60)
base_soup = BeautifulSoup(base_source, 'html.parser')
```

```python
all_author = base_soup.find_all(class_='box yf-anchor anchor')
```

```python
all_author[0]
```

```python
authors_rank = []
authors_list = []
authors_link = []
authors_volgers = []
authors_favorieten = []

for i in range(0, len(all_author)):
    
    authors_rank.append(i + 1)
    authors_list.append(all_author[i]['title'])
    authors_link.append(all_author[i]['data-url'])
    
    print(str(i + 1), ") ", all_author[i]['title'], all_author[i]['data-url'])
    
    prop = all_author[i].find_all(class_='property')
    
    volgers = []
    favorieten = []
    
    for p in prop:
        if "volgers" in p.get_text():
            volgers.append(p.get_text())
        if "favorieten" in p.get_text():
            favorieten.append(p.get_text())
            
    if not volgers:
        volgers.append('no data')
        
    if not favorieten:
        favorieten.append('no data')
        
    authors_volgers.append(volgers[0])
    authors_favorieten.append(favorieten[0])
    
    print("**********************************************************************************")
```

```python
authors_base_data = pd.DataFrame(list(zip(authors_rank, authors_list, authors_link, authors_volgers, authors_favorieten)),
                            columns =['position', 'author_name', 'link', 'volgers', 'favorieten'])

authors_base_data['volgers'] = authors_base_data['volgers'].str.extract('(^\d*)')
authors_base_data['favorieten'] = authors_base_data['favorieten'].str.extract('(^\d*)')

authors_base_data
```

```python
authors_base_data.to_csv('authors_toplist.csv', index=False)
```

```python
author_book_link = []

authors_gelezen = []
authors_aanhetlezen = []
authors_wiliklezen = []
authors_favoriet = []
authors_recensies = []
authors_artikelen = []
authors_newsroom = []
authors_spots = []
authors_topics = []

i = 1

for book_link in authors_base_data['link']:
    
    print(i)
    
    author_book_link.append(book_link)
    
    driver.get(book_link)
    time.sleep(2)
    base_source_book = driver.page_source
    base_soup_book = BeautifulSoup(base_source_book, 'html.parser')
    print(book_link)
    
    author_stats = base_soup_book.find_all(class_='author-stats')
    
    author_rank = []
    author_gelezen = []
    author_aanhetlezen = []
    author_wiliklezen = []
    author_favoriet = []
    author_recensies = []
    
    for stats in author_stats:
        
        stat_label = stats.find(class_='author-stats-label')
        stat_value = stats.find(class_='author-stats-value')
        
        stat_label_s = stat_label.get_text()        
        stat_value_s = stat_value.get_text()
        
        
        if stat_label_s == 'Rank':
            author_rank.append(stat_value_s)
        
        if stat_label_s == 'Gelezen':
            author_gelezen.append(stat_value_s)
            
        if stat_label_s == 'Aan het lezen':
            author_aanhetlezen.append(stat_value_s)
            
        if stat_label_s == 'Wil ik lezen':
            author_wiliklezen.append(stat_value_s)
            
        if stat_label_s == 'Favoriet':
            author_favoriet.append(stat_value_s)
            
        if stat_label_s == 'Recensies':
            author_recensies.append(stat_value_s)
    
    if not author_rank:
        author_rank.append('no data')
    
    if not author_gelezen:
        author_gelezen.append('no data')
    
    if not author_aanhetlezen:
        author_aanhetlezen.append('no data')
    
    if not author_wiliklezen:
        author_wiliklezen.append('no data')
    
    if not author_favoriet:
        author_favoriet.append('no data')
    
    if not author_recensies:
        author_recensies.append('no data')
        
    authors_rank.append(author_rank[0])
    authors_gelezen.append(author_gelezen[0])
    authors_aanhetlezen.append(author_aanhetlezen[0])
    authors_wiliklezen.append(author_wiliklezen[0])
    authors_favoriet.append(author_favoriet[0])
    authors_recensies.append(author_recensies[0])
    
    author_stats2 = base_soup_book.find_all(class_='btn btn4')
    
    author_artikelen = []
    author_newsroom = []
    author_spots = []
    author_topics = []
    
    for stats2 in author_stats2:
        
        stats2_text = stats2.get_text()
        stats2_text = stats2_text.replace('\n', '')
        stats2_text = stats2_text.replace('\t', '')
        
        if 'Artikelen' in stats2_text:
            author_artikelen.append(stats2_text)
        if 'Newsroom' in stats2_text:
            author_newsroom.append(stats2_text)
        if 'Spots' in stats2_text:
            author_spots.append(stats2_text)
        if 'Topics' in stats2_text:
            author_topics.append(stats2_text)
            
    if not author_artikelen:
        author_artikelen.append('no data')
        
    if not author_newsroom:
        author_newsroom.append('no data')
    
    if not author_spots:
        author_spots.append('no data')
    
    if not author_topics:
        author_topics.append('no data')
        
    authors_artikelen.append(author_artikelen[0])
    authors_newsroom.append(author_newsroom[0])
    authors_spots.append(author_spots[0])
    authors_topics.append(author_topics[0])
    
    i = i + 1
```

```python
authors_extra_data = pd.DataFrame(list(zip(author_book_link,
                                          authors_gelezen,
                                          authors_aanhetlezen,
                                          authors_wiliklezen,
                                          authors_favoriet,
                                          authors_recensies,
                                          authors_artikelen,
                                          authors_newsroom,
                                          authors_spots,
                                          authors_topics)),
                                 
                            columns =['link',
                                      'gelezen',
                                      'aanhetlezen',
                                      'wiliklezen',
                                      'favoriet',
                                     'recensies',
                                     'artikelen',
                                     'newsroom',
                                     'spots',
                                     'topics'])

def find_number(text):
    num = re.findall(r'[0-9]+',text)
    return(" ".join(num))

authors_extra_data['gelezen'] = authors_extra_data['gelezen'].str.extract('(^\d*)')
authors_extra_data['aanhetlezen'] = authors_extra_data['aanhetlezen'].str.extract('(^\d*)')
authors_extra_data['wiliklezen'] = authors_extra_data['wiliklezen'].str.extract('(^\d*)')
authors_extra_data['favoriet'] = authors_extra_data['favoriet'].str.extract('(^\d*)')
authors_extra_data['recensies'] = authors_extra_data['recensies'].str.extract('(^\d*)')
authors_extra_data['artikelen']=authors_extra_data['artikelen'].apply(lambda x: find_number(x))
authors_extra_data['newsroom']=authors_extra_data['newsroom'].apply(lambda x: find_number(x))
authors_extra_data['spots']=authors_extra_data['spots'].apply(lambda x: find_number(x))
authors_extra_data['topics']=authors_extra_data['topics'].apply(lambda x: find_number(x))


authors_extra_data
```

```python
authors_all_data = pd.merge(authors_base_data, authors_extra_data, how='left', left_on='link', right_on='link')
```
